```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_merge = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_merge.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data_merge.columns
```

```python
total_schools=len(school_data_merge["school_name"].value_counts())

total_students=school_data_merge["Student ID"].count()

total_budget=school_data_merge["budget"].sum()

ave_math=school_data_merge["math_score"].mean()

ave_reading=school_data_merge["reading_score"].mean()

overall_pass_score=(ave_math + ave_reading)/2

math_pass_num=school_data_merge.loc[school_data_merge["math_score"] >= 70, ["math_score"]]
math_pass_percentage =math_pass_num['math_score'].count()/ total_students


reading_pass_num=school_data_merge.loc[school_data_merge["reading_score"] >= 70, ["reading_score"]]
reading_pass_percentage=reading_pass_num['reading_score'].count()/total_students

```

```python
district_summary_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget":[total_budget],
    "Average Math Score": [ave_math],
    "Average Reading Score": [ave_reading],
    "Overall Pass Score": [overall_pass_score],
    "% Passing Math": [math_pass_percentage],
    "% Passing Reading": [reading_pass_percentage]
})


district_summary_df.style.format({
    "Total Budget": "${:,.2f}", 
    "Average Reading Score": "{:.1f}", 
    "Average Math Score": "{:.1f}", 
    "% Passing Math": "{:.1%}", 
    "% Passing Reading": "{:.1%}", 
    "Overall Passing Rate": "{:.1%}"
})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_type = school_data.set_index("school_name")["type"]
```

```python
school_student = school_data_merge["school_name"].value_counts()
```

```python
school_budget=school_data.groupby(["school_name"]).sum()["budget"]
```

```python
per_student_budget= school_budget / school_student
```

```python
school_ave_math=school_data_merge["math_score"].mean()

school_ave_reading=school_data_merge["reading_score"].mean()
```

```python
mathpass_percent= school_data_merge[school_data_merge["math_score"]>70].groupby(["school_name"])
mathpass_percent = mathpass_percent["math_score"].count()/school_student

readpass_percent= school_data_merge[school_data_merge["reading_score"]>70].groupby(["school_name"])
readpass_percent = readpass_percent["reading_score"].count()/school_student

overall_pass=(mathpass_percent+ readpass_percent)/2
```

```python
school_summary = pd.DataFrame({

    "School Type": school_type,
    "Total Students": school_student,
    "Total Budget":school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": school_ave_math,
    "Average Reading Score": school_ave_reading,
    "% Passing Math": mathpass_percent,
    "% Passing Reading": readpass_percent,
    "% Overall Passing Rate": overall_pass
})

school_summary.style.format({'Total Students': '{:,}', 
                           "Total School Budget": "${:,}", 
                           "Per Student Budget": "${:.0f}",
                           'Average Math Score': "{:.1f}", 
                           'Average Reading Score': "{:.1f}", 
                           "% Passing Math": "{:.1%}", 
                           "% Passing Reading": "{:.1%}", 
                           "Overall Passing Rate": "{:.1%}"})
```

## Math Scores by Grade

  * Group each series by school
  
  * Combine the series into a dataframe

```python
nineth_grade = school_data_merge[(school_data_merge["grade"] == "9th")]
tenth_grade = school_data_merge[(school_data_merge["grade"] == "10th")]
eleventh_grade = school_data_merge[(school_data_merge["grade"] == "11th")]
twelfth_grade = school_data_merge[(school_data_merge["grade"] == "12th")]

nineth_math = nineth_grade.groupby(["school_name"]).mean()["math_score"]
tenth_math = tenth_grade.groupby(["school_name"]).mean()["math_score"]
eleventh_math = eleventh_grade.groupby(["school_name"]).mean()["math_score"]
twelfth_math = twelfth_grade.groupby(["school_name"]).mean()["math_score"]

math_scores_by_grade = pd.DataFrame({"9th Grade": nineth_math, 
                                     "10th Grade": tenth_math,
                                     "11th Grade": eleventh_math,
                                     "12th Grade": twelfth_math})

math_scores_by_grade = math_scores_by_grade[["9th Grade", "10th Grade", "11th Grade", "12th Grade"]]

math_scores_by_grade
```

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_summary["Per Student Budget"] = pd.cut(per_student_budget, spending_bins, labels=group_names)
```

```python
school_spending_summary = pd.DataFrame({"Average Math Score": spending_ave_math,
                                                  "Average Reading Score": spending_ave_read,
                                                  "% Passing Math": spending_pass_math,
                                                  "% Passing Reading": spending_pass_read, 
                                                  "Overall Passing Rate": spending_overall})



school_spending_summary.style.format({'Total Students': '{:,}', 
                           "Total School Budget": "${:,}", 
                           "Per Student Budget": "${:.0f}",
                           'Average Math Score': "{:.1f}", 
                           'Average Reading Score': "{:.1f}", 
                           "% Passing Math": "{:.1%}", 
                           "% Passing Reading": "{:.1%}", 
                           "Overall Passing Rate": "{:.1%}"})

```
